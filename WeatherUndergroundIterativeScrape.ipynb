```python
# a command to import URLs
import requests

# get Beautiful Soup (already installed)
from bs4 import BeautifulSoup

#create a new text file that can store the results of the script
f = open("wunder-data.txt" , "w")

# this entire bit is to scrape a year's worth of max temperatures, one day/page at a time from weather underground

# this tells us how many months, days to go through
for m in range(1,13):
    for d in range(1,32):
        if(m==2 and d>28):
            break
        elif(m in[4,6,9,11] and d>30):
            break
        
        # this part controls how the data is written to the file
        timestamp = "2010" + str(m) + str(d)
        print ("Getting data for " + timestamp)
        url = "https://www.wunderground.com/history/airport/KSFO/2010/" + str(m) + "/" + str(d) + "/DailyHistory.html"
        
        # we use requests.get to pull the data from the URL above and save that as r
        r = requests.get(url)
        
        # this is the scraping part. It is looking within the soup for a particular value "span" and then within that for "wx-value" and then for the first (0) time it appears.
        # The soup is r (from previous line) read as text
        soup = BeautifulSoup(r.text)
        dayTemp = soup.find_all("span", "wx-value")[0].string
        
        #writes the months as two digits (so jan is 01 not 1)
        if len(str(m))<2:
            mStamp = "0" + str(m)
        else:
            mStamp = str(m)
            
        #writes the days as two digits
        if len(str(d))<2:
            dStamp = "0" +str(d)
        else:
            dStamp = str(d)
            
        timestamp = "2010" + mStamp + dStamp
        
        #writes scraped data to the file
        f.write(timestamp + "," + dayTemp + "\n")
        
        #closes the file
        f.close()
```
